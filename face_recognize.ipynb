```python
# Install the required libraries
!pip install opencv-python face_recognition

# Import libraries
import cv2
import face_recognition

# Load the uploaded image
import numpy as np
from PIL import Image
from io import BytesIO

# Upload the image to Google Colab
from google.colab import files
uploaded = files.upload()

# Convert the uploaded image to a format suitable for OpenCV
image_path = list(uploaded.keys())[0]
img = Image.open(BytesIO(uploaded[image_path]))
img_np = np.array(img)

# Use face_recognition to detect faces in the image
known_image = face_recognition.load_image_file(image_path)  # Load the image
known_face_encoding = face_recognition.face_encodings(known_image)[0]

# Draw a rectangle around the detected face
face_locations = face_recognition.face_locations(known_image)
for (top, right, bottom, left) in face_locations:
    cv2.rectangle(img_np, (left, top), (right, bottom), (0, 255, 0), 2)

# Display the image with the rectangle around the face
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(img_np, cv2.COLOR_BGR2RGB))
plt.show()
```
